제주특별자치도의 성별 분포를 항아리 모양 그래프로 표현

```python
import csv
import matplotlib.pyplot as plt
```

```python
import csv

f = open('/content/gender.csv', encoding='cp949')
data = csv.reader(f)

male = []
female = []

for row in data:
  if '제주특별자치도' in row[0]:
    for i in row[3 : 104]:
      male.append(-int(i.replace(',','')))
    for i in row[106 : ] :
      female.append(int(i.replace(',','')))
    break
    

plt.figure(dpi=200)
plt.title('제주특별자치도 지역의 남녀 성별 인구 분포')
plt.barh(range(101), male, color = 'skyblue',label = '남성')
plt.barh(range(101), female, color = 'hotpink', label = '여성')
plt.legend()
plt.show()
```

```python
print(len(male),len(female))
```

항아리 모양의 그래프는 연령대별 인구 차이 확인 가능 

but 어떤 성별이 더 많은지에 대한 정보 한눈에 파악하기 어려움 -> 파이차트 Pie


```python
plt.figure(dpi=200)
plt.pie([10,20]) #pie(x) x가 가지고있는 값을 비율로 표시
plt.show()
```

```python
plt.figure(dpi=200)
size=[2441,2312,1031,1233]
label = ['A형','B형','AB형','O형']
plt.axis('equal')
plt.pie(size,labels=label,autopct='%.1f%%') #혈액형의 비율을 임의로 작성
plt.legend()
plt.show()
```

```python
plt.figure(dpi=200)
size=[2441,2312,1031,1233]
label = ['A형','B형','O형','AB형']
color = ['blue','deeppink','hotpink','pink']
plt.axis('equal')
plt.pie(size,labels=label,autopct='%.1f%%',explode=(0,0,0,0.2),colors=color) #혈액형의 비율을 임의로 작성
plt.legend()
plt.show()
```

제주도의 성별 인구 비율 표현하기

```python
import csv

f = open('/content/gender.csv', encoding='cp949')
data = csv.reader(f)

male = []
female = []

size=[] #제주도의 남성 인구 합계, 여성 인구 합계

for row in data:
  if '세종특별자치시' in row[0]:
    male = 0    #남성 인구수 합계 변수
    female = 0  #여성 인구수 합계 변수
    for i in range(101):
      male += int(row[3+i].replace(',',''))  
      female += int(row[106+i].replace(',',''))
    break

size.append(male)
size.append(female)

#파이차트 시각화

plt.figure(dpi=200)
label = ['남자','여자']
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size,labels=label,autopct='%.1f%%',startangle=90,explode=(0,0),colors=color) #혈액형의 비율을 임의로 작성
plt.legend()
plt.show()
```
