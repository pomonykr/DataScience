```python
import csv

f = open('/content/201902_201902_연령별인구현황_월간.csv',encoding='cp949')
data = csv.reader(f)

result = []

#지역명을 입력받는 변수 선언
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')

for row in data:
  if name in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(',','')))

#인구구조 시각화
plt.figure(dpi=200)
plt.style.use('ggplot')  #격자 무늬 스타일로 지정
plt.plot(result)
plt.show()
```

항아리 모양 그래프 그리기

성별데이터 저장하기(type1)

```python
import csv

f = open('/content/201901_201901_연령별인구현황_월간.csv',encoding='cp949')
data = csv.reader(f)

m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in range(0,101):
      m.append(int(row[i+3]))
      f.append(int(row[-(i+1)]))



print(f)
f.reverse() #0세 부터 인구수 나올 수 있도록 데이터 뒤집기
print(f)
```

성별 데이터 순서대로 저장하기(Type2)

```python
import csv

f = open('/content/201901_201901_연령별인구현황_월간.csv',encoding='cp949')
data = csv.reader(f)

m = []
f = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3 : 104]:
      m.append(int(i))
    for i in row[106 : ] :
      f.append(-int(i))

print(m)
print(f)
```

데이터 시각화 하기

```python
plt.figure(dpi=200)
plt.rc('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label='여성')
plt.legend()
plt.show()
```

2. 2019년 2월 연령별 인구구조 -> 2020년 2월 연령별 인구 구조 비교(같이 그리기)

```python
import csv

f = open('/content/gender.csv', encoding='cp949')
data = csv.reader(f)

male = []
female = []

for row in data:
  if '고운동' in row[0]:
    for i in row[3 : 104]:
      male.append(int(i))
    for i in row[106 : ] :
      female.append(-int(i))
    

plt.figure(dpi=200)
plt.title('고운동 지역의 남녀 성별 인구 분포')
plt.barh(range(101), male, color = 'skyblue',label = '남성')
plt.barh(range(101), female, color = 'hotpink', label = '여성')
plt.legend()
plt.show()
```
