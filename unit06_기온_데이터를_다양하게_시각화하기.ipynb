1월과 8월의 데이터를 히스토그램으로 시각화하기

```python
import csv
import matplotlib.pyplot as plt


f = open('/content/seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data) #헤더

janTemp = [] #1월의 최고 기온을 담은 aug 리스트 변수 선언

augTemp = [] #8월의 최고 기온을 담은 aug 리스트 변수 선언

for row in data:
  month = row[0].split('-')[1] #날짜 데이터에서 월 데이터 추출
  
  if row[-1] != '':
    if month == '08' : #8월 데이터면 aug 변수에 담기
      augTemp.append(float(row[-1]))
    if month == '01' : #1월 데이터면 jan 변수에 담기
      janTemp.append(float(row[-1]))

plt.figure(dpi=200)
plt.hist(janTemp,bins=100,color='b',label='Aug')
plt.hist(augTemp,bins=100,color='r',label='Jan')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


f = open('/content/seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data) #헤더

janTemp = [] #1월의 최고 기온을 담은 aug 리스트 변수 선언
juneTemp = []
augTemp = [] #8월의 최고 기온을 담은 aug 리스트 변수 선언

for row in data:
  month = row[0].split('-')[1] #날짜 데이터에서 월 데이터 추출
  
  if row[-1] != '':
    if month == '08' : #8월 데이터면 aug 변수에 담기
      augTemp.append(float(row[-1]))
    if month == '04' : #8월 데이터면 aug 변수에 담기
      juneTemp.append(float(row[-1]))
    if month == '01' : #1월 데이터면 jan 변수에 담기
      janTemp.append(float(row[-1]))

plt.figure(dpi=200)
plt.hist(janTemp,bins=100,color='b',label='Aug')
plt.hist(juneTemp,bins=100,color='g',label='Jun')
plt.hist(augTemp,bins=100,color='r',label='Jan')
plt.legend()
plt.show()
```

1월부터 12월까지 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data) #헤더

month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1])) # 월과 같은 번호의 인덱스에 월별 데이터 저장 (예: 1월 row[0].split('-')[1] -> month[0] row[0].split('-')[1]-1 , 2월 -> month[1])


plt.figure(dpi=200)
plt.boxplot(month)
plt.show()
```

8월 1일 부터 31일까지 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data) #헤더

day = []
for i in range(31) :
  day.append([])

for row in data:
  if row[-1] !='':
    if row[0].split('-')[1] == '08': # 8월 최고 기온 값 저장 day[0] -> 1일
      day[int(row[0].split('-')[2])-1].append(float(row[-1])) 

plt.style.use('ggplot') # 그래프 스타일 지정 (회색 배경의 격자무늬, 2/4 값을 의미하는 선의 색 변경)
plt.figure(dpi=200)
plt.boxplot(day,showfliers=False) #아웃라이어 이상치 값 생략
plt.show()
```

2) 상자그림을 활용한 최고 기온 vs 서울 최고 기온


```python
import csv
import matplotlib.pyplot as plt

f = open('/content/seoul.csv',encoding='cp949')
d = open('/content/busan.csv',encoding='cp949')
data1 = csv.reader(f)
data2 = csv.reader(d)
next(data1) #헤더
next(data2)

seoul = [] #서울 최고 기온 데이터를 저장할 리스트 생성
busan = [] #부산 최고 기온 데이터를 저장할 리스트 생성
 
for row1 in data1:
  if row1[-1] != '':
    seoul.append(float(row1[-1]))
for row2 in data2:
  if row2[-1] != '':
    busan.append(float(row2[-1]))

plt.figure(dpi=200)
plt.boxplot([seoul,busan],labels=['seoul','busan']) #아웃라이어 이상치 값 생략
plt.legend()
plt.show()

```

3) 월별 최고기온 상자 그림 그리기


```python
import csv
import matplotlib.pyplot as plt

f = open('/content/busan.csv',encoding='cp949')
data = csv.reader(f)
next(data) #헤더

month = []
for i in range(12):
  month.append([])

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1])) # 월과 같은 번호의 인덱스에 월별 데이터 저장 (예: 1월 row[0].split('-')[1] -> month[0] row[0].split('-')[1]-1 , 2월 -> month[1])


plt.figure(dpi=200)
plt.boxplot(month,showfliers=False)
plt.show()
```
