꺽은선 그래프로 표현하기

```python
import csv

f = open('/content/gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

for row in data:
  if '제주특별자치도' in row[0]:
    for i in row[3 : 104]:
      m.append(int(i.replace(',','')))
    for i in row[106 : ] :
      f.append(int(i.replace(',','')))
    break
    
plt.style.use('ggplot')
plt.figure(dpi=200)
plt.rc('font',family='NanumBarunGothic')
#꺾은선 그래프로 남여 인구수 데이터 표현하기
plt.plot(m,label='남성')
plt.plot(f,label='여성')
plt.legend()
plt.show()
```

연령별로 남성 인구수에서 여성 인구수ㅜ를 뺀 차이의 막대그래프 그리기


```python
import csv

f = open('/content/gender.csv', encoding='cp949')
data = csv.reader(f)

result = []

for row in data:
  if '서울특별시' in row[0]:
    for i in range(3,104):
      result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
    break

plt.style.use('ggplot')
plt.figure(dpi=200)
plt.rc('font',family='NanumBarunGothic')
#꺾은선 그래프로 남여 인구수 데이터 표현하기
plt.rcParams['axes.unicode_minus'] = False
plt.title('지역의 남녀 성별인구 차이 분포')
plt.bar(range(101),result)
plt.show()
```

위치, 크기가 서로 다른 100개의 점 만들기

```python
import random
x = []
y = []
size = []

for i in range(100) :
  x.append(random.randint(50,100))
  y.append(random.randint(50,100))
  size.append(random.randint(10,100))

plt.figure(dpi=200)
plt.style.use('ggplot')
plt.scatter(x,y, s=size) #scatter(x,y)
plt.colorbar()
plt.show() 
```

```python
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.scatter(x,y, s=size,cmap='jet',c=size) #scatter(x,y)
plt.colorbar()
plt.show()
```

```python
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.scatter(x,y, s=size,cmap='jet',c=size,alpha=0.7) # alpha 투명도 추가 0~1: 0에 가까울수록 투명, 1에 가까울수록 불투명
plt.colorbar()
plt.show()
```

원의 크기는 인구수의 제곱근 추가

```python
import csv
import math

f = open('/content/gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []
size = []

for row in data:
  if '제주특별자치도' in row[0]:
    for i in row[3 : 104]:
      m.append(int(i.replace(',','')))
      size.append(math.sqrt(int(i.replace(',',''))))
    for i in row[106 : ] :
      f.append(int(i.replace(',','')))
    break
    
plt.figure(dpi=200)
#산점도 그래프로 남여 인구수 데이터 표현하기
plt.scatter(m,f,s=size,c=range(101),cmap='jet',alpha=0.5) #x축은 남자,y축은 여자
plt.plot(range(max(m)),range(max(m)),'g')
plt.colorbar()
plt.show()
```
