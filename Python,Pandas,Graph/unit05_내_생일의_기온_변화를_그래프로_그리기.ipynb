데이터 준비하기

```python
#최고 기온 데이터만 출력하기
```

```python
import csv

f = open('/content/seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data) #헤더

result = [] #최고 기온 데이터를 저장할 리스트 생성

for row in data:
  if row[-1] != '':
    result.append(float(row[-1]))

print(result) #최고 기온 리스트
```

매년 생일 기준 최고 기온, 최저 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt


f = open('/content/seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data) #헤더

high = []
low  = []

for row in data:
  if row[-1] != '' and row[-2] != '':
    if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '17':
      high.append(float(row[-1])) #최고기온
      low.append(float(row[-2])) #최저기온

plt.figure(dpi=200)  
plt.plot(high,'r',label='high')
plt.plot(low,'b',label='low')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


f = open('/content/seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data) #헤더

high = []
low  = []

# 무려 1987년 12월 14일생

for row in data:
  if row[-1] != '' and row[-2] != '':
    if 1987 <= int(row[0].split('-')[0]):
      if row[0].split('-')[1]  == '12' and row[0].split('-')[2] == '14':
        high.append(float(row[-1])) #최고기온
        low.append(float(row[-2]))

plt.figure(dpi=200)  
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.plot(high,'r',label='high')
plt.plot(low,'b',label='low')
plt.title('내 생일의 기온 변화 그래프')
plt.legend() #범례 표시
plt.show()
```

매년 생일 기준 최고 기온, 평균 기온, 최저 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt


f = open('/content/seoul.csv',encoding='cp949')
data = csv.reader(f)
next(data) #헤더

high = []
avg  = []
low  = []


# 무려 1987년 12월 14일생

for row in data:
  if row[-1] != '' and row[-2] != '' and row[-3] != '':
    if 1987 <= int(row[0].split('-')[0]):
      if row[0].split('-')[1]  == '12' and row[0].split('-')[2] == '14':
        high.append(float(row[-1])) #최고기온
        avg.append(float(row[-3]))
        low.append(float(row[-2]))


plt.figure(dpi=200)  
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.plot(high,'r',label='high')
plt.plot(avg,'green',label='avg')
plt.plot(low,'b',label='low')
plt.title('내 생일의 기온 변화 그래프')
plt.legend() #범례 표시
# plt.hist([1,1,2,3,4,5,6,6,7,8,10])
plt.show()
```
