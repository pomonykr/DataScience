plt.plot (데이터셋) 그래프 그리기

```python
import matplotlib.pyplot as plt  #라이브러리 임포트
```

```python
plt.plot([10,20,30,40])      # y 값이 주어지면, 자동으로 x값을 0부터 1씩 증가함
plt.show()
```

plt.plot (x,y)그래프 그리기

```python
plt.plot([1,2,3,4],[12,43,25,15])
plt.show
```

```python
plt.title('plotting')
plt.plot([1,2,3,4],[12,43,25,15])
plt.show
```

```python
plt.title('plotting')

plt.plot([10,20,30,40],label='asc') #증가를 의미하는 asc 범례
plt.plot([40,30,20,10],label='desc') #감소를 의미하는 desc 범례

plt.legend(loc=5) #그래프에 범례 속성 추가
plt.show()
```

```python
plt.title('plotting')

plt.plot([10,20,30,40],color='r', label='red') #증가를 의미하는 asc 범례 #기본색상은 약자로 가능
plt.plot([40,30,20,10],'k',label='black') #감소를 의미하는 desc 범례

plt.legend(loc=5) #그래프에 범례 속성 추가
plt.show()
```

```python
plt.title('linestyle')


plt.plot([10,20,30,40],'r.',label='circle') #증가를 의미하는 asc 범례 #기본색상은 약자로 가능
plt.plot([40,30,20,10],'g^',label='triangle up') #감소를 의미하는 desc 범례

plt.legend() #그래프에 범례 속성 추가
plt.show()
```

```python
plt.title('marker')

#'r.--' : <색상><마커모양><선모양> 코드로 작성 가능


plt.plot([10,20,30,40],'r.-',label='circle') #증가를 의미하는 asc 범례 #기본색상은 약자로 가능
plt.plot([40,30,20,10],'g^',label='triangle up') #감소를 의미하는 desc 범례

plt.legend() #그래프에 범례 속성 추가
plt.show()
```

데이터 시각화하기

```python
import csv
```

```python
f = open('/content/seoul.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data) #헤더 추출

result = []  #결측치를 제거한 최고 기온 리스트

for row in data :
  if row[-1] != '' :   #최고기온이 ''(빈문자열)이 아니면,
    result.append(float(row[-1]))
    
result

plt.figure(dpi=200)
plt.plot(result,'b')
plt.show()
```
